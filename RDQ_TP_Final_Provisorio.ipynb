```python
#Empiezo a normalizar la informacion
#PASO 1 Normalizar fechas en los datasets
mis_df = [df_clientes, df_marketing, df_ventas]
for df in mis_df:
  for columna in df.columns:
    if "fecha" in columna.lower():
      df[columna] = pd.to_datetime(df[columna], errors="coerce", dayfirst=True)
```

```python
#Para verificar que los tipos esten bien ahora
print(df_clientes.dtypes)
print(df_marketing.dtypes)
print(df_ventas.dtypes)
```

```python
#PASO 2 Limpiar las todos los dataframe. Normalizar el texto el general
#Primero armo una lista de los dataframes a normalizar
#y a cada uno de elos los recorro solo por sus columnas del tipo "object"
#los transformo a String para trabajarlos mas facil, les quito espacios vacios
#les quito caracteres de espacios vacios, saltos de linea, tabulaciones, etc
#Finalmente normalizo todos los textos con las primeras letras en mayuscula
mis_df = [df_clientes, df_marketing, df_ventas]
for df in mis_df:
  for columna in df.select_dtypes(include="object").columns:
    df[columna] = df[columna].astype(str)
    df[columna] = df[columna].str.strip()
    df[columna] = df[columna].str.replace(r"[\u200b\t\r\n]", "", regex=True)
    df[columna] = df[columna].str.replace(r"\s+", " ", regex=True)
    df[columna] = df[columna].str.title()
```

### 2. Aplicar filtros y transformaciones para crear una tabla de ventas que muestre los productos con alto rendimiento

```python
#Empiezo con la primera transformacion de datos. Genero el Subtotal (precio * cantidad) sobre el
#DataFrame Ventas
#df_ventas["subtotal"] = df_ventas["precio"] * df_ventas["cantidad"]
df_ventas = df_ventas.assign(subtotal = df_ventas["precio"] * df_ventas["cantidad"])
df_ventas.head(10)
```

```python
#Creo un nuevo DataFrame tomado de df_ventas agrupado por producto y sumando los subtotales
df_rendimiento = df_ventas.groupby(by="producto", dropna=False, as_index=False)
df_rendimiento = df_rendimiento.agg(unidades = ("cantidad", "sum"), precio_promedio = ("precio", "mean"), registros = ("precio", "size"), total = ("subtotal", "sum"))
#Redondeo los decimales del precio_promedio
df_rendimiento["precio_promedio"] = df_rendimiento["precio_promedio"].round(2)
#Muestro los productos que mayor recaudacion generaron en orden descendente
df_rendimiento.sort_values(by= "total", ascending=False, ignore_index=True)
```

```python
#Para averiguar el percentil 80 del dataframe anterior
p80 = df_rendimiento["total"].quantile(0.8, interpolation="linear")
p80 = p80.round(2)
print("El percentil 80 de todos nuestros ingresos es:", p80, "\n")

#Armo un nuevo DataFrame con el listado de los productos de Alto Rendimiento (arriba del percentil 80)
ventas_p80 = df_rendimiento.query("total >= @p80", engine="python")
ventas_p80 = ventas_p80.sort_values(by= "total", ascending=False, ignore_index=True)
print("PRODUCTOS DE ALTO RENDIMIENTO (20% que generan mayores ingresos)")
print(ventas_p80)
```
