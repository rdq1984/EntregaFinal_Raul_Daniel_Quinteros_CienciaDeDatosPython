```python
#Empiezo a normalizar la informacion
#PASO 1 Normalizar fechas en los datasets
mis_df = [df_clientes, df_marketing, df_ventas]
for df in mis_df:
  for columna in df.columns:
    if "fecha" in columna.lower():
      df[columna] = pd.to_datetime(df[columna], errors="coerce", dayfirst=True)
```

```python
#Para verificar que los tipos esten bien ahora
print(df_clientes.dtypes)
print(df_marketing.dtypes)
print(df_ventas.dtypes)
```

```python
#PASO 2 Limpiar las todos los dataframe. Normalizar el texto el general
#Primero armo una lista de los dataframes a normalizar
#y a cada uno de elos los recorro solo por sus columnas del tipo "object"
#los transformo a String para trabajarlos mas facil, les quito espacios vacios
#les quito caracteres de espacios vacios, saltos de linea, tabulaciones, etc
#Finalmente normalizo todos los textos con las primeras letras en mayuscula
mis_df = [df_clientes, df_marketing, df_ventas]
for df in mis_df:
  for columna in df.select_dtypes(include="object").columns:
    df[columna] = df[columna].astype(str)
    df[columna] = df[columna].str.strip()
    df[columna] = df[columna].str.replace(r"[\u200b\t\r\n]", "", regex=True)
    df[columna] = df[columna].str.replace(r"\s+", " ", regex=True)
    df[columna] = df[columna].str.title()
```

### 2. Aplicar filtros y transformaciones para crear una tabla de ventas que muestre los productos con alto rendimiento

```python
#Empiezo con la primera transformacion de datos. Genero el Subtotal (precio * cantidad) sobre el
#DataFrame Ventas
#df_ventas["subtotal"] = df_ventas["precio"] * df_ventas["cantidad"]
df_ventas = df_ventas.assign(subtotal = df_ventas["precio"] * df_ventas["cantidad"])
df_ventas.head(10)
```

```python
#Creo un nuevo DataFrame tomado de df_ventas agrupado por producto y sumando los subtotales
df_rendimiento = df_ventas.groupby(by="producto", dropna=False, as_index=False)
df_rendimiento = df_rendimiento.agg(unidades = ("cantidad", "sum"), precio_promedio = ("precio", "mean"), registros = ("precio", "size"), total = ("subtotal", "sum"))
#Redondeo los decimales del precio_promedio
df_rendimiento["precio_promedio"] = df_rendimiento["precio_promedio"].round(2)
#Muestro los productos que mayor recaudacion generaron en orden descendente
df_rendimiento.sort_values(by= "total", ascending=False, ignore_index=True)
```

```python
#Para averiguar el percentil 80 del dataframe anterior
p80 = df_rendimiento["total"].quantile(0.8, interpolation="linear")
p80 = p80.round(2)
print("El percentil 80 de todos nuestros ingresos es:", p80, "\n")

#Armo un nuevo DataFrame con el listado de los productos de Alto Rendimiento (arriba del percentil 80)
df_ventas_p80 = df_rendimiento.query("total >= @p80", engine="python")
df_ventas_p80 = df_ventas_p80.sort_values(by= "total", ascending=False, ignore_index=True)
print("PRODUCTOS DE ALTO RENDIMIENTO (el 20% que generan mayores ingresos)")
print(df_ventas_p80)
```

### 3. Resumir las ventas por categoria de producto y analizar los ingresos generados

```python
#Realizo un nuevo DataFrame donde agrupo por categoria sobre el dataframe ventas
df_ventasxcate = df_ventas.groupby(by="categoria", dropna=False, as_index=False)
#y aplico funciones de agregacion para los parametros cantidad y subtotal
df_ventasxcate = df_ventasxcate.agg(unidades_vendidas = ("cantidad", "sum"), total_vendido = ("subtotal", "sum"))
print("TOTAL DE VENTAS POR CATEGORIAS")
display(df_ventasxcate)
```

### 4. Integración de datos, OPCIONAL

```python
# TODO: Combinar los sets de datos de ventas y marketing para obtener una visión más amplia de las tendencias.
# Genere un nuevo DataFrame uniendo las ventas de mayor rendimiento, para ver cuanto fue el costo total
# del marketing para los productos mas vendidos
df_marketingtotal = df_marketing.groupby(["producto"], as_index=False)["costo"].sum()
df_ventas_marketing = pd.merge(df_ventas_p80, df_marketingtotal, on="producto", how="inner")
#display(df_ventas_marketing)
#Y finalmente vamos a calcular el costo porcentual del marketing aplicado a las ventas de estos productos
df_ventas_marketing['%_costo'] = (df_ventas_marketing['costo'] / df_ventas_marketing['total']) * 100
display(df_ventas_marketing)

# hago lo mismo para comprar contra el 20% que menos venta genero
# Tambien calculo el costo % del marketing para los productos menos vendidos
p20 = df_rendimiento["total"].quantile(0.2, interpolation="linear")
p20 = p20.round(2)
df_ventas_p20 = df_rendimiento.query("total < @p20", engine="python")
df_ventas_p20 = df_ventas_p20.sort_values(by= "total", ascending=False, ignore_index=True)
df_ventas_marketing2 = pd.merge(df_ventas_p20, df_marketingtotal, on="producto", how="inner")
df_ventas_marketing2['%_costo'] = (df_ventas_marketing2['costo'] / df_ventas_marketing2['total']) * 100
display(df_ventas_marketing2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y=df_ventas_marketing['%_costo'])
plt.title('Boxplot del Porcentaje de Costo sobre Ingreso de los Productos Mas Vendidos')
plt.ylabel('Porcentaje de Costo sobre Ingreso')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(y=df_ventas_marketing2['%_costo'])
plt.title('Boxplot del Porcentaje de Costo sobre Ingreso de los Productos Menos Vendidos')
plt.ylabel('Porcentaje de Costo sobre Ingreso')
plt.show()
```

1. Estadística Descriptiva: Calcular medidas de tendencia central y dispersión sobre las ventas. Presentar los resultados en un informe.

```python
# Grafico de barras para mostrar la distribucion de las ventas por producto

df_grafico1 = df_rendimiento.copy()
df_grafico1 = df_grafico1.sort_values("total", ascending=False)
plt.figure(figsize=(14,10))
color_verde ="#09C87C"
color_azul = "#272AF5"
color_rojo = "#F54927"
barra = plt.bar(df_grafico1["producto"], df_grafico1["total"], color=color_verde)
plt.title("Ingresos Totales por Producto", fontsize=14, weight="bold")
plt.xlabel("Productos", fontsize=12)
plt.ylabel("Ingreso Total ($)", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# MEDIDAS DE TENDENCIA CENTRAL
# Ya los habiamos calculado en la primera parte del TP, pero los volvemos a calcular, ahora sobre
# el DataFrame de Ventas por Producto
media = df_grafico1['total'].mean()
mediana = df_grafico1['total'].median()
moda = df_grafico1['total'].mode()

print("MEDIDAS DE TENDENCIA CENTRAL")
print("Media:", media)
print("Mediana:", mediana)
print("Moda:\n", moda)
```

```python
# MEDIDAS DE DISPERSION (Rango, Varianza, Desviacion Estandar)
# Calculamos las medidas de dispersion del mismo DataFrame

rango = df_grafico1['total'].max() - df_grafico1['total'].min()
varianza = df_grafico1['total'].var(ddof=1)
desviacion = df_grafico1['total'].std(ddof=1)

print(f"Media: ${media:,.2f}")
print(f"Mediana: ${mediana:,.2f}")
print(f"Rango: ${rango:,.2f}")
print(f"Varianza: {varianza:,.2f}")
print(f"Desviacion Estandar: ${desviacion:,.2f}")

print("CONCLUSION: Encontramos que el promedio (o media) la cual esta afectada por valores outliers es 48,903.12")
print("y la mediana es 48,140.24, por lo que son valores muy cercanos, por lo que podemos entender en un primer vistazo")
print("que todos los valores deben ser bastante simetricos, sin que los valores extremos la modifiquen significativamente")
print("La desviacion estandar (cuanto se alejan los valores del promedio) es de 13,265.56, aproximadamente un 27% de la media")
```

```python
q1 = df_grafico1['total'].quantile(0.25)
q2 = df_grafico1['total'].quantile(0.5)
q3 = df_grafico1['total'].quantile(0.75)
iqr = q3 - q1

print("RANGO INTERCUARTILICO")
print(f"Q1: ${q1:,.2f}")
print(f"Mediana (Q2): ${q2:,.2f}")
print(f"Q3: ${q3:,.2f}")
print(f"IQR: ${iqr:,.2f}")
```

```python
# DETECCCION DE OUTLIERS CON EL IQR
# Usamos las formulas vistas para listar los valores outliers

limite_inf = q1 - 1.5 * iqr
limite_sup = q3 + 1.5 * iqr
outliersiqrsup = df_grafico1[(df_grafico1['total'] > limite_sup)]
outliersiqrinf = df_grafico1[(df_grafico1['total'] < limite_inf)]
print("DETECCION DE OUTLIERS CON IQR")
print(f"Limite Inferior: {limite_inf:,.2f}")
print(f"Limite Superior: {limite_sup:,.2f}")
print("\nOutliers Superiores:")
display(outliersiqrsup)
print("\nOutliers Inferiores:")
display(outliersiqrinf)
```
